```python
import numpy as np
import matplotlib.pyplot as plt

def create_sphere (rho, number_points, error = 0.0, plot_flag = True):
    theta1 = np.linspace (0,2 * np.pi, round(np.sqrt(number_points)))
    theta2 = np.linspace (0,    np.pi, round(np.sqrt(number_points)))
    points = []
    for i in theta1:
        for j in theta2:
            rho_rand = rho + rho * error * np.random.randn(1) 
            new_point = list([rho_rand* np.sin(j) * np.cos(i), rho_rand * np.sin(j) * np.sin(i), rho_rand * np.cos(j)])
            points.append(new_point) 
    points = np.asarray(points)

    fig = plt.figure()
    ax = fig.add_subplot(projection= '3d')
    ax.scatter(points[:,0],points[:,1],points[:,2], s = 0.5)

    return points[:,:,0]
```

```python
points = create_sphere (rho = 1, number_points = 1000, error = 0.01, plot_flag = True)
np.shape(points)
```

```python
np.savetxt("file.xyz", points, delimiter=" ", header="Numero di atomi\n3", comments="")
```
