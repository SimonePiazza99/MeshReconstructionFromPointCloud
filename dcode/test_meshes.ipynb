```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def generate_points (range_x, mean, std, flag_remove= False, plot_flag = True):
    points = []
    for i in np.linspace(range_x[0],range_x[1], range_x[1]-range_x[0]+1):
        y_rand = np.random.normal(mean, std)
        points.append([i, y_rand])
    if flag_remove:
        points.pop(int((range_x[1]-range_x[0]+1)/2))
    points = np.asarray(points)
    if plot_flag:
        plt.scatter(points[:,0], points[:,1], c='black', zorder = 10)
    return points
```

```python
# Definisci le coordinate dei punti del piano
range_x = [-5,5]
x = np.linspace(range_x[0], range_x[1], 100)  # Coordinate x
retta1 = np.full(len(x),1)
retta2 = np.full(len(x),-1)
# Disegna il piano
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, retta1, c = 'lightblue',linewidth = 3 , label ='surface 1')
ax.plot(x, retta2, c = 'lightgreen', linewidth = 3, label ='surface 2')
ax.plot([range_x[0], range_x[0]],[-1,1], c = 'darkgreen', linewidth = 3)
ax.axis('equal')  # Imposta gli assi uguali

#punti
punto_orignale = [0,1]
plt.scatter(punto_orignale[0], punto_orignale[1], c = 'red', s = 50, zorder = 10)

ax.quiver(0, 1, 0, 1, angles='xy', scale_units='xy', scale=1, color='black')

ax.legend()

points1 = generate_points(range_x, 1, 0.0003, flag_remove= True)
points2 = generate_points(range_x, -1, 0.0003)
```
