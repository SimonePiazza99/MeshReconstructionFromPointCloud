```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Genera un cilindro con asse lungo l'asse x
radius = 0.5  # Raggio del cilindro
height = 5.0  # Altezza del cilindro
num_points_circ = 10  # Numero di punti nella nuvola
num_points_height = 20

# Genera i punti sulla superficie del cilindro
theta = np.linspace(0, 2 * np.pi, num_points_circ)
x = np.linspace(0, height, num_points_height)

theta, x = np.meshgrid(theta, x)
z = radius * np.cos(theta)
y = radius * np.sin(theta)

# crea dataset x y z
x = x.flatten()
y = y.flatten()
z = z.flatten()

# Seleziona un punto con z massima e x = 1.0
eps = 0.27
idx = np.where((z == np.max(z)) & (height/2 - eps < x ) & (x < height/2 + eps ))


# Plotta la nuvola di punti
fig = plt.figure()  
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='grey', alpha = 0.3)

#Plotta contorno cilindro in trasparenza
# ax.plot_surface(x.reshape(num_points_height, num_points_circ),
#                  y.reshape(num_points_height, num_points_circ), 
#                  z.reshape(num_points_height, num_points_circ), alpha=0.4)

# # Plotta contorno cilindro in colore solido con bordo
# ax.plot_surface(x.reshape(num_points_height, num_points_circ),
#                 y.reshape(num_points_height, num_points_circ), 
#                 z.reshape(num_points_height, num_points_circ), 
#                 alpha=0.4, color='blue', edgecolor='black')

# # Imposta l'illuminazione
# ax.light_sources = [(1, 1, 1)]

# # Allinea la vista
# ax.view_init(elev=30, azim=45)

# # Zoomma la vista
# ax.dist = 8

# Nascondi griglia
# ax.axis('off')
# Seleziona un punto con z massima:

print(idx)
idx = 99
# Plotta la nuvola di punti)
ax.scatter(x[idx], y[idx], z[idx], c='red', alpha=1.0)

# crea freccia che esce da x[idx], y[idx], z[idx]
ax.quiver(x[idx], y[idx], z[idx], 0.0, 0.0, 1.0, length=1.0, normalize=True, color='black')

# Crea cono che abbia come asse la freccia con vertice in x[idx], y[idx], z[idx]
# Calcola le coordinate per il cono quadrico
cone_height = 1.0  # Altezza del cono
cone_radius = np.tan(np.radians(45)) * cone_height  # Raggio del cono
cone_points = np.array([x[idx], y[idx], z[idx]])

# Crea una griglia di punti per il cono quadrico
u = np.linspace(0, 2 * np.pi, 30)
v = np.linspace(-cone_height, cone_height, 30)
u, v = np.meshgrid(u, v)

# Calcola le coordinate x, y, z per il cono quadrico
x_cone = cone_points[0] + (v * cone_radius * np.cos(u))
y_cone = cone_points[1] + (v * cone_radius * np.sin(u))
z_cone = cone_points[2] + v

# Plotta il cono quadrico
ax.plot_surface(x_cone, y_cone, z_cone, color='red', alpha=0.5)


# Allinea la vista
ax.view_init(elev=25, azim = 45)

# Zoomma la vista
ax.dist = 5

```

```python
# PROVA CIRC
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parametri della semisfera
radius = 1.0  # Raggio della semisfera
num_points = 20  # Numero di punti nella nuvola

# Genera i punti sulla semisfera
u = np.linspace(0, np.pi/2, num_points)
v = np.linspace(0, 2 * np.pi, num_points)
u, v = np.meshgrid(u, v)
x = radius * np.sin(u) * np.cos(v)
y = radius * np.sin(u) * np.sin(v)
z = radius * np.cos(u)

# Plotta la nuvola di punti
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='grey')

# visualizza solo quelli con z > 0.8
idx = np.where(z > 0.7)
ax.scatter(x[idx], y[idx], z[idx], c='red')

# Rimuovi il box
# ax.axis('off')

# Imposta gli assi
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Allinea la vista sul piano XY
ax.view_init(elev=0, azim=45)

# Mostra il grafico
plt.show()


```
